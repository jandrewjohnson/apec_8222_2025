### Spatial coordinate references

- Just like with Rasters, these have a CRS:
  - It is saved as an attribute on the GDF.

```
# Print the CRS
gdf.crs
```

### Change the CRS

- Let's create a more advanced plot that illustrates the different projections we might use

```
proj_names = ["WGS 84", "Mercator", "Winkel Tripel", "Quartic Authalic"]
crs_names = ["EPSG:4326", "EPSG:3395", "+proj=wintri", "+proj=qua_aut"]

fig, axes = plt.subplots(
    nrows=2, ncols=2, figsize=(12, 6), dpi=150, sharex=False, sharey=False
)
for i, ax in enumerate(axes.flat):
    gdf.to_crs(crs_names[i]).plot(ax=ax, color="#b2f3b2", linewidth=0.1)
    ax.set_title(proj_names[i])
    ax.axis("off")
plt.tight_layout()
plt.show()
```

### Get the centroid

- For better or worse, almost all of "spatial economics" ignores vector data
  - Instead, if there is space involved, it uses the centroid of the polygon
  - This will be used for, e.g., calculating distance between a village and a forest
- We can get the centroids out of our GDF and plot them

```
gdf["centroid"] = gdf.centroid

fig, ax = plt.subplots()
ax.axis("off")
gdf.boundary.plot(ax=ax, lw=0.2, color="k")
gdf["centroid"].plot(ax=ax, color="red", markersize=15)
plt.show()
```

# Or Buffers

```
# Create a buffer around each centroid (e.g. radius = 0.1 degrees or map units)
centroid_buffers = gdf.centroid.buffer(3)

# Make a new GeoDataFrame for convenience
buffer_gdf = gpd.GeoDataFrame(gdf[['nev_name']], geometry=centroid_buffers, crs=gdf.crs)

buffer_gdf.plot()
```

# Common Connection Back to Spatial Economics

Suppose you have a village represented as a point (centroid) and you want to know the area within 5 km of that village that is forested. You can create a buffer around the centroid point representing a 5 km radius, and then analyze the land use within that buffer area to determine how much of it is forested. This technique is commonly used in spatial economics to assess the impact of proximity to resources or services on economic outcomes.
